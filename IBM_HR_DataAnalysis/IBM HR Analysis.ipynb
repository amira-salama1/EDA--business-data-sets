### Data Cleaning

> This dataset is pretty Clean no duplicates or null values, datatypes are correct, however Attrition variable should change  to be boolean value 

```python
# Making a copy of the new df
df_n = df.copy()
```

```python
# Checking Attrition Values
df_n.Attrition.value_counts()
```

```python
#Changing attrition values to be 1 for yes and zero if no
df_n['Attrition'] = df_n['Attrition'].apply(lambda x : 1 if x == 'Yes' else 0)
```

```python
#Test
df_n['Attrition'].value_counts()
```

## Univariate Exploration


```python
#plotting count plot for attrition variable
sb.countplot(data=df, x='Attrition');
print(df.Attrition.value_counts(normalize=True))
```

> Attrition rate of the company is 16 %.

```python
#Plotting a histogram of the distance from home variable
plt.hist(data=df_n, x='DistanceFromHome') 
plt.title("Distribution of the Distance From Home")
plt.xlabel("Distance From Home");
```

> The distribution of the distance from home is right skewed where most of the data lies nearest to 5 miles.

```python
sb.displot(df.TotalWorkingYears, bins = 20);
print(df.TotalWorkingYears.value_counts(bins=3))
```

> The distribution of Total working years is also right skewed, most of the values are between 5 and 13 years

```python
#Plotting a bar chart for job involvement
color =sb.color_palette()[0]
sb.countplot(data=df, x = 'JobInvolvement', color=color)
plt.xlabel('Job Involvement');
```

> The highest Job Involvement rate is 3 which is a high Job Involvement level.

```python
#Plotting a bar chart for Work Life Balance
sb.countplot(data=df, x = 'WorkLifeBalance', color=color)
plt.xlabel('Work Life Balance');
```

> 3 means a very good level of Work Life Balance, and it has the highest value.

```python
sb.countplot(data=df, x ='Gender')
print(df.Gender.value_counts(normalize=True));
```

> Male employees are 60 % of the company employment, females make the remaining 40%.

```python
#plotting departments in a horizontal barplot
df.Department.value_counts().plot(kind = 'barh')
plt.title('Count of Employees by Department')
plt.ylabel("Department");
```

> Research and Dev. has the most employees the comes sales and Human Resources

```python
#plotting age ditribution among employees
sb.displot(df.Age);
```

```python
df.Age.value_counts(normalize=True, bins=5)
```

> The Age variable seem to be normally distributed, most of the ages of the employees lies between 30 and 43

```python
# HISTOGRAM for Years In Current Role
plt.figure(figsize=(10,4))
plt.subplot(1, 3, 1)
plt.hist(data=df, x='YearsInCurrentRole',bins=20);
plt.xlabel('Years In Current Role')

# HISTOGRAM for Years Since Last Promotion
plt.subplot(1,3, 2)
plt.hist(data=df, x='YearsSinceLastPromotion', bins=20)
plt.xlabel('Years Since Last Promotion')

# HISTOGRAM for Years With Current Manager
plt.subplot(1,3,3)
plt.hist(data=df, x='YearsWithCurrManager', bins=20)
plt.xlabel('Years With Current Manager');
```
