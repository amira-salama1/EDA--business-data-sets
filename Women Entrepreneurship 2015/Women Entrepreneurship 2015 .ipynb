```python
df_female = pd.DataFrame(df_femme[['Country', 'Women Entrepreneurship Index']])
df_female.sort_values('Women Entrepreneurship Index', inplace=True, ascending = False)
df_female[:10]
```

```python
# Showing graph of Women Entrepreneurship Index by Country in Decending order

fig, ax = plt.subplots(figsize = (12,12))


ax.barh(df_female.Country, df_female['Women Entrepreneurship Index'], align='center')
ax.set_yticks(df_female.Country)
ax.set_yticklabels(df_female.Country)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_title('Women Entrepreneurship Index by Country')

plt.show()
```
